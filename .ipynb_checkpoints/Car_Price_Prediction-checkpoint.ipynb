```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
cars = pd.read_csv("CarPrice_Assignment.csv")
cars.head()
```

```python
print("Information of data:\n", cars.info())
print("Size of data:\n", cars.shape)
print("describtion of data:\n", cars.describe())
```

```python
# Checking the uniqueness in the categorical values in columns
print("Fuel Type:  ",cars.fueltype.unique())
print("Aspiration: ",cars.aspiration.unique())
print("NO.of Doors:",cars.doornumber.unique())
print("Car Body:   ",cars.carbody.unique())
print("Type Wheel: ",cars.drivewheel.unique())
print("Engine Loc: ",cars.enginelocation.unique())
print("Engine Typ: ",cars.enginetype.unique())
print("Cyc No:     ", cars.cylindernumber.unique())
print("Fuel sys :  ",cars.fuelsystem.unique())
```

```python
print(cars.drivewheel.value_counts())
# Here we can see "4wd" is the error in data As 4 and r can be miss type we are converting "4wd" to "rwd"
cars.drivewheel.replace("4wd","rwd",inplace=True)
print(cars.drivewheel.value_counts())
```

```python
cars["car_company"],cars["car_model"] = cars["CarName"].str.split(" ",1).str
cars.head()
```

```python
# Converting the categorical to numeric
cars["fueltype"] = cars["fueltype"].map({"gas": 0, "diesel": 1})  # GAS == 0 , DIESEL == 1
cars["aspiration"] = cars["aspiration"].map({"std":0,"turbo":1})   # STD == 0 , TURBO == 1
cars["doornumber"] =cars["doornumber"].map({"two":2,"four":4})     # TWO == 2 , FOUR == 4
cars["drivewheel"] = cars["drivewheel"].map({"rwd":0,"fwd":1})     # rwd == 0 , fwd == 1
cars["enginelocation"] = cars["enginelocation"].map({"front":1,"rear":0})  # front == 1 , rear == 0
cars["cylindernumber"] = cars["cylindernumber"].map({"four":4,"six":6,"five":5,"three":3,"twelve":12,"two":2,"eight":8})
```

```python
cars.cylindernumber.unique()
```

```python
# So know we have all unique values in the categorical data
# Drop the carid column
cars.drop(columns=["car_ID","CarName"],inplace=True)
cars.head()
```

```python
# Visualizing the data
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.boxplot(x="carbody",y="price",data=cars)
plt.subplot(1,2,2)
sns.boxplot(x="enginetype",y="price",data=cars)
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.boxplot(x="cylindernumber",y="price",data=cars)
plt.subplot(1,2,2)
sns.boxplot(x="fuelsystem",y="price",data=cars)
plt.show()
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(cars.corr(), annot = True, cmap="YlGnBu")
plt.show()
```
