```python
# imporinting the required libraries
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#reading the data file
cars = pd.read_csv("CarPrice_Assignment.csv")
cars.head()
```

```python
# analysis the breif data
print("Information of data:\n", cars.info())
print("Size of data:\n", cars.shape)
print("describtion of data:\n", cars.describe())
```

```python
#spilting the CarName into car_company and car_model
cars["car_company"],cars["car_model"] = cars["CarName"].str.split(" ",1).str
cars.drop(columns=["CarName"])
cars.head()
```

```python
print(cars.drivewheel.value_counts())
# Here we can see "4wd" is the error in data As 4 and r can be miss type we are converting "4wd" to "rwd"
cars.drivewheel.replace("4wd","rwd",inplace=True)
print(cars.drivewheel.value_counts())
```

```python
cars.columns.unique()
```

```python
# Get Dummies for catogarical data
#Fueltype
fuel_type_status = pd.get_dummies(cars["fueltype"],drop_first=True)
cars = pd.concat([cars,fuel_type_status],axis=1)
#aspiration
aspiration_status = pd.get_dummies(cars["aspiration"],drop_first=True)
cars = pd.concat([cars,aspiration_status],axis=1)
#doornumber
door_number_status = pd.get_dummies(cars["doornumber"],drop_first=True)
cars = pd.concat([cars,door_number_status],axis=1)
#carbody
car_body_status = pd.get_dummies(cars["carbody"],drop_first=True)
cars = pd.concat([cars,car_body_status],axis=1)
#drivewheel
drive_wheel_status = pd.get_dummies(cars["drivewheel"],drop_first=True)
cars = pd.concat([cars,drive_wheel_status],axis=1)
#enginelocation
engine_location_status = pd.get_dummies(cars["enginelocation"],drop_first=True)
cars = pd.concat([cars,engine_location_status],axis=1)
#enginetype
engine_type_status = pd.get_dummies(cars["enginetype"],drop_first=True)
cars = pd.concat([cars,engine_type_status],axis=1)
#cylindernumber
cylinder_number_status = pd.get_dummies(cars["cylindernumber"],drop_first=True)
cars = pd.concat([cars,cylinder_number_status],axis=1)
#fuelsystem
fuel_system_status = pd.get_dummies(cars["fuelsystem"],drop_first=True)
cars = pd.concat([cars,fuel_system_status],axis=1)
#company_name
car_company_status = pd.get_dummies(cars["car_company"],drop_first=True)
cars = pd.concat([cars,car_company_status],axis=1)
```
